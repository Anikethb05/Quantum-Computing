#### 1. Density Matrix for Bell State

```python
QC=QuantumCircuit(2)
QC.h(0)
QC.cx(0,1)
QC.draw('mpl')
```

```python
state = Statevector.from_instruction(QC)
print("Statevector =",state)
```

```python
rho_bell_state=DensityMatrix(state)
```

```python
plot_state_city(rho_bell_state)
```

#### Plot the Density Matrices for the following:

#### 1. 0.8|0><0| + 0.2|1><1|  (mixed state)
#### 2. (|0> + i|1>)/√2 (pure state with complex phase)


```python
#1.
rho_0=DensityMatrix.from_label('0')
rho_1=DensityMatrix.from_label('1')

rho=0.8*rho_0 + 0.2*rho_1
plot_state_city(rho)
```

```python
#2.
# |ψ> = (|0> + i|1>)/√2
psi = Statevector([1/2**0.5, 1j/2**0.5])

# Compute density matrix
rho = psi.to_operator().data
print(rho)
```

```python
plot_state_city(rho)
```

#### State Construction - Create a Density Matrix for a mixed state that is:

#### - 60% probability of being in |0>
#### - 40% probability of being in |1>



```python
rho_0=DensityMatrix.from_label('0')
rho_1=DensityMatrix.from_label('1')

rho=0.6*rho_0 + 0.4*rho_1
plot_state_city(rho)
```

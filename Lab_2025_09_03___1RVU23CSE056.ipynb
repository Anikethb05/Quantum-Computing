### 2. The SWAP operator

```python
QC = QuantumCircuit(2)
```

```python
# q0 = 1/2|0> + sqrt(3)/2|1> and q1 = |1>

QC.ry(2*(pi)/3, 0)
QC.x(1)
```

```python
State1 = Statevector(QC)
plot_bloch_multivector(State1)
```

```python
# SWAP - 3 alternate CNOTs
QC.cx(0,1)
QC.cx(1,0)
QC.cx(0,1)
```

```python
State2 = Statevector(QC)
plot_bloch_multivector(State2)
```

### 5. Demonstrating the Toffoli Gate (CCNOT)

```python
QC = QuantumCircuit(3)
```

```python
QC.x(0)  # Set q0 = 1 (control 1 ON)
QC.x(1)  # Set q1 = 1 (control 2 ON)

QC.ccx(0,1,2) # Toffoli Gate -- Output : 111
```

```python
QC.measure_all()
job = Simulator.run(QC)
results = job.result().get_counts()
QC.draw('mpl')
```

```python
plot_histogram ( results , title = " Aniketh Bhargav (056) " )
```

```python
QC = QuantumCircuit(3)
QC.x(0)  # Set q0 = 1 (control 1 ON)
# Set q1 = 0 (control 2 OFF)
```

```python
QC.ccx(0,1,2) # Toffoli Gate  -- Output: 100
```

```python
QC.measure_all()
job = Simulator.run(QC)
results = job.result().get_counts()
QC.draw('mpl')
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector
```

```python
simulator=AerSimulator()
```

## AND

```python
QC=QuantumCircuit(3)
QC.h([0,1])
QC.ccx(0,1,2)
QC.measure_all()
```

```python
QC.draw('mpl')
```

```python
result=simulator.run(QC).result()
count=result.get_counts()
plot_histogram(count)
```

## XOR

```python
QC=QuantumCircuit(3)

QC.h([0,1])
QC.cx(0,2)
QC.cx(1,2)
QC.measure_all()
```

```python
QC.draw('mpl')
```

```python
result=simulator.run(QC).result()
count=result.get_counts()
plot_histogram(count)
```

## OR

```python
QC=QuantumCircuit(3)

QC.h([0,1])
QC.cx(0,2)
QC.cx(1,2)
QC.ccx(0,1,2)
QC.measure_all()
```

```python
QC.draw('mpl')
```

```python
result=simulator.run(QC).result()
count=result.get_counts()
plot_histogram(count)
```

## Half-Adder

```python
QC=QuantumCircuit(3)

QC.h([0,1])
QC.ccx(0,1,2)
QC.cx(0,1)
```

```python
QC.draw('mpl')
```

```python
QC.measure_all()

result=simulator.run(QC).result()
count=result.get_counts()
plot_histogram(count)
```

## Full-Adder

```python
QC=QuantumCircuit(4)

QC.h([0,1,2])

QC.ccx(0,1,3)
QC.ccx(0,2,3)
QC.ccx(1,2,3)
QC.cx(0,2)
QC.cx(1,2)

QC.measure_all()
```

```python
QC.draw('mpl')
```

```python
result=simulator.run(QC).result()
count=result.get_counts()
plot_histogram(count)
```

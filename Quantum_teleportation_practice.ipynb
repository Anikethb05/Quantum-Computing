```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector
import numpy as np
```

```python
QC=QuantumCircuit(3,3)
```

```python
QC.initialize([1/np.sqrt(3), np.sqrt(2)/np.sqrt(3)], 0)
```

```python
state0= Statevector(QC)
```

```python
state0.draw('latex')
```

```python
QC.h(1)
QC.cx(1,2)
QC.draw()
```

```python
QC.draw('mpl')
```

```python
state1=Statevector(QC)
```

```python
state1.draw('latex')
```

```python
QC.cx(0,1)
QC.h(0)
```

```python
QC.draw('mpl')
```

```python
state2=Statevector(QC)
```

```python
state2.draw('latex')
```

```python
QC.measure([0,1],[0,1])
```

```python
simulator=AerSimulator()
```

```python
result=simulator.run(QC, shots=1024).result()
counts=result.get_counts()
plot_histogram(counts)
```

```python
with QC.if_test( (QC.clbits[1],1)):
  QC.x(2)
```

```python
with QC.if_test( (QC.clbits[0],1)):
  QC.z(2)
```

```python
QC.measure(2,2)
```

```python
QC.draw('mpl')
```

```python
result1=simulator.run(QC).result()
count1=result1.get_counts()
plot_histogram(count1)
```

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
```

```python
def create_oracle_instruction(oracle_type):
    oracle_circuit = QuantumCircuit(2)
    if oracle_type == 'constant_0':
        oracle_circuit.id(1)
    elif oracle_type == 'constant_1':
        oracle_circuit.x(1)
    elif oracle_type == 'balanced_0':
        oracle_circuit.cx(0, 1)
    elif oracle_type == 'balanced_1':
        oracle_circuit.cx(0, 1)
        oracle_circuit.x(1)
    else:
        raise("oracle_type should be one of 'constant_0', 'constant_1', 'balanced_0' or 'balanced_1'")
    return oracle_circuit.to_instruction(label='f-box')
```

```python
Simulator=AerSimulator()
```

```python
for oracle_type in ['constant_0', 'constant_1', 'balanced_0', 'balanced_1']:
    QC056 = QuantumCircuit(2, 1)
    QC056.x(1)
    QC056.h(0)
    QC056.h(1)
    QC056.barrier()

    oracle_instruction = create_oracle_instruction(oracle_type)
    QC056.append(oracle_instruction, [0, 1])
    QC056.barrier()
    QC056.h(0)
    QC056.measure(0, 0)

    # Transpile and run
    transpiled_QC = transpile(QC056, Simulator)
    job = Simulator.run(transpiled_QC)
    result = job.result()
    counts = result.get_counts()

    # âœ… Display circuit and histogram inline in Colab
    display(QC056.draw('mpl'))
    display(plot_histogram(counts, title=f"Oracle Type: {oracle_type}"))
```

```python
# circuit 1: constant_0
QC056 = QuantumCircuit(2, 1)

QC056.x(0)
QC056.x(1)

QC056.h(0)
QC056.h(1)

QC056.barrier()

# Apply the custom oracle instruction
oracle_instruction = create_oracle_instruction('constant_0')
QC056.append(oracle_instruction, [0, 1])

QC056.barrier()


QC056.h(0)
QC056.draw('mpl')
```

```python
state = Statevector(QC056)
plot_bloch_multivector(state)
```

```python
from qiskit import transpile

QC056.measure(0,0)
Simulator=AerSimulator()

# Transpile the circuit to the simulator's basis gates
transpiled_qc = transpile(QC056, Simulator)

job=Simulator.run(transpiled_qc)
result=job.result()
counts=result.get_counts()
plot_histogram(counts)
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_state_city
from qiskit.quantum_info import Statevector, DensityMatrix, partial_trace
import numpy as np
```

```python
def calc_purity(rho):
  m=rho.data@rho.data
  purity=np.trace(m)
  return purity
```

```python
rho_0=DensityMatrix.from_label('0')
rho_1=DensityMatrix.from_label('1')
```

```python
rho=0.3*rho_0+0.7*rho_1
purity=calc_purity(rho)
print(purity)
```

```python
plot_state_city(rho)
```

```python
purity_0=calc_purity(rho_0)
purity_1=calc_purity(rho_1)
print(purity_0,purity_1)
```

```python
builtin_0=rho_0.purity()
builtin_1=rho_1.purity()
```

```python
print(builtin_0, builtin_1)
```

```python
QC=QuantumCircuit(2)
QC.h(0)
QC.cx(0,1)
```

```python
QC.draw('mpl')
```

```python
state=Statevector.from_instruction(QC)
```

```python
rho=DensityMatrix(state)
```

```python
plot_state_city(rho)
```

```python
print(calc_purity(rho))
print(rho.purity())
```

```python
rho_0=partial_trace(rho,[1])
rho_1=partial_trace(rho,[0])
plot_state_city(rho_1)
```

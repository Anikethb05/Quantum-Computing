```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import matplotlib.pyplot as plt
import numpy as np
```

```python
QC=QuantumCircuit(3,3)
QC.draw('mpl')
```

```python
QC.initialize([1/np.sqrt(3),np.sqrt(2)/np.sqrt(3)], 0)
```

```python
QC.draw('mpl')
```

```python
state0 = Statevector(QC)
```

```python
state0.draw(output='latex')
```

```python
plot_bloch_multivector(state0)
```

```python
QC.h(1)
QC.cx(1,2)
```

```python
QC.draw('mpl')
```

```python
state1 = Statevector(QC)
```

```python
state1.draw(output='latex')
```

```python
plot_bloch_multivector(state1)
```

```python
QC.cx(0,1)
QC.h(0)
```

```python
QC.draw('mpl')
```

```python
state2 = Statevector(QC)
```

```python
state2.draw(output='latex')
```

```python
plot_bloch_multivector(state2)
```

```python
QC.measure([0,1],[0,1])
```

```python
Simulator=AerSimulator()
```

```python
job=Simulator.run(QC)
result=job.result()
counts=result.get_counts()
```

```python
plot_histogram(counts)
```

```python
with QC.if_test( (QC.clbits[1], 1) ):
  QC.x(2)
```

```python
with QC.if_test( (QC.clbits[0], 1) ):
  QC.z(2)
```

```python
QC.measure(2,2)
```

```python
QC.draw('mpl')
```

```python
job=AerSimulator().run(QC)
result=job.result()
counts=result.get_counts()
```

```python
plot_histogram(counts)
```
